# saving results

## test

```python
columns = ["num_stages", "min_alt_freq", "num_alt", "max_alt_per_pathway", "seed", "num_strictly_consistent_patients", 
           "num_partially_consistent_patients", "time_took_problem", "time_took_other", "avg_alterations_per_pathway", 
           "num_null_datasets", "p1_value", "effect_size"]

AML_results_file_path = os.path.join(ff.get_file_path("project_root"), "final", "results", "AML")

patient_trees = ff.get_patient_trees_lists("AML")
num_stages_values = [2]
min_alt_freq_values = [3]
max_alt_per_pathway_values = [3]
num_null_datasets = 1000
seed = 1

results = [] # gonna be list of lists

combos = list(itertools.product(num_stages_values, min_alt_freq_values, max_alt_per_pathway_values))
for num_stages, min_alt_freq, max_alt_per_pathway in tqdm(combos):
    
    save_folder_path = os.path.join(AML_results_file_path, f"{num_stages}_stages_{min_alt_freq}_min_alt_freq_{max_alt_per_pathway}_mapp")
    num_pathways = num_stages
    num_dependencies = num_stages - 1

    results.append(ff.run_problem_1_1_and_save_result(patient_trees, min_alt_freq, num_stages, num_pathways, num_dependencies, max_alt_per_pathway, num_null_datasets, seed, save_folder_path))

    seed += 1

df = pd.DataFrame(results, columns=columns)

df.to_csv(os.path.join(AML_results_file_path, "summary.csv"), index=False)
```
